```python
import pandas as pd
import numpy
from pathlib import Path
import numpy as np
from torchvision.utils import make_grid
import torch
import matplotlib.pyplot as plt
```

```python
# Define data paths
train_csv = '../data/train.csv'
test_csv = '../data/test.csv'
```

# Explore data

```python
train_df = pd.read_csv(train_csv)
test_df = pd.read_csv(test_csv)
```

```python
num_train_examples = len(train_df)
num_features = train_df.columns
labels = train_df.label
labels_set = set(labels)


print('Number of training examples: {}'.format(num_train_examples))
print('Number of features: {}'.format(len(num_features) - 1))
print('Number of classes: {}'.format(len(labels_set)))
print('Label values: {}'.format(labels_set))
```

```python
num_test_examples = len(test_df)
num_features = len(test_df.columns)

print('Number of test examples: {}'.format(num_test_examples))
print('Number of features: {}'.format(num_features))
```

## Visualize random images

```python
# .values
selections = np.random.randint(num_train_examples, size=10)
visual_data = train_df.iloc[selections, 1:].values.reshape(-1, 1, 28, 28)
visual_data_ts = torch.from_numpy(visual_data)
```

```python
visual_grid = make_grid(visual_data_ts, nrow=5)
plt.imshow(visual_grid.permute(1, 2, 0))
```

## Data distribution

```python
plt.rcParams['figure.figsize'] = (18,5)
freqs = train_df.label.value_counts(sort=False)
plt.bar(freqs.index, freqs)
plt.xticks([i for i in range(10)])
```
